```python
import os
import numpy as np
from cellpose import core, models, io

use_GPU = core.use_gpu()
yn = ['NO', 'YES']
print(f'>>>GPU activated? {yn[use_GPU]}')


# default parameters
diameter = 0
flow_threshold = 0.4
cellprob_threshold = 0
chan = 0
chan2 = 0
#

neuron_model_path = '/home/dell/Documents/LeGAO/Task_NucleiSegmentation/new_0716/a_training_1101/train_set_3_1101/neuron/models/cellpose_residual_on_style_on_concatenation_off_neuron_2023_11_01_15_07_25.697110'
neuron_model = models.CellposeModel(gpu=use_GPU, pretrained_model=neuron_model_path)

neuron_diameter = neuron_model.diam_labels if diameter == 0 else diameter


glia_model_path = '/home/dell/Documents/LeGAO/Task_NucleiSegmentation/new_0716/a_training_1101/train_set_3_1101/glia/models/cellpose_residual_on_style_on_concatenation_off_glia_2023_11_01_15_29_41.908358'
glia_model = models.CellposeModel(gpu=use_GPU, pretrained_model=glia_model_path)

glia_diameter = glia_model.diam_labels if diameter == 0 else diameter

```

```python
import pandas as pd

df = pd.read_csv('/home/dell/Documents/LeGAO/Task_NucleiSegmentation/new_0716/CodePython/path_slice_51.txt')

NumSlice = len(df["DirPath"])

print(df["DirPath"][0])
print(df['DirPath'][0]+'/neuron')
```

```python
## check file exist
for iSlice in range(NumSlice):
    print(os.path.exists(df["DirPath"][iSlice]))
    print(os.path.exists(df["FilePath"][iSlice]))
```

```python
for iSlice in range(NumSlice):
    # read image
    tmp_file_name=df['FilePath'][iSlice]
    print(tmp_file_name)

    tmp_image=io.imread(tmp_file_name)

    #segment neuron
    neuron_masks, neuron_flows, neuron_styles = neuron_model.eval(
        x=tmp_image,
        channels=[chan, chan2],
        diameter=neuron_diameter,
        flow_threshold=flow_threshold,
        cellprob_threshold=cellprob_threshold
    )

    # create dir
    neuron_dir = df['DirPath'][iSlice]+'/neuron_v2'

    isExist = os.path.exists(neuron_dir)
    if not isExist:
        os.makedirs(neuron_dir)
        print('Create neuron dir')


    io.save_masks(
        tmp_image, 
        neuron_masks, 
        neuron_flows, 
        tmp_file_name,
        channels=[chan, chan2],
        png=False,
        tif=True,
        save_txt=True,
        save_flows=False,
        save_outlines=False,
        savedir=neuron_dir
        )
    

    #segment glia
    glia_masks, glia_flows, glia_styles = glia_model.eval(
        x=tmp_image,
        channels=[chan, chan2],
        diameter=glia_diameter,
        flow_threshold=flow_threshold,
        cellprob_threshold=cellprob_threshold
    )

    # create dir
    glia_dir = df['DirPath'][iSlice]+'/glia_v2'

    isExist = os.path.exists(glia_dir)
    if not isExist:
        os.makedirs(glia_dir)
        print('Create glia dir')


    io.save_masks(
        tmp_image, 
        glia_masks, 
        glia_flows, 
        tmp_file_name,
        channels=[chan, chan2],
        png=False,
        tif=True,
        save_txt=True,
        save_flows=False,
        save_outlines=False,
        savedir=glia_dir
        )
```
